```python
import matplotlib.pyplot as plt
import numpy as np
import math
import random

dataset  = [[1.5,2.5,"Bad Student"], [2.3,2,"Bad Student"], [3.2,2.7,"Bad Student"], [7.2,8,"Good Student"], [8,6,"Good Student"], [9,7,"Good Student"]]

for i in range(100):
    r1 = random.randint(1,100)
    r1 = r1/10
    r2 = random.randint(1,100)
    r2 = r2/10
    if ((r1+r2)/2) < 5:
        dataset.append([r1, r2, "Bad Student"])
    else:
        dataset.append([r1, r2, "Good Student"])
    

x1 = []
y1 = []

x2 = []
y2 = []

testP1 = [1,1]
testP2 = [9,9]

in1 = int(input("val1: "))
in2 = int(input("val2: "))
new = [in1, in2]

boo = []
haa = []

minimum = math.inf
res = None
for i in dataset:
    z = [i[0], i[1]]
    vec1 = np.subtract(z, testP1)
    vec2 = np.subtract(z, testP2)
    l1 = math.sqrt((vec1[0]**2 + vec1[1]**2))
    l2 = math.sqrt((vec2[0]**2 + vec2[1]**2))
    if l1 > l2:
        boo.append(z)
    else:
        haa.append(z)
        
for i in boo:
    x1.append(i[0])
    y1.append(i[1])

for i in haa:
    x2.append(i[0])
    y2.append(i[1])


plt.plot(x1,y1, 'yo')
plt.plot(x2,y2, 'mo')
plt.plot(testP1[0],testP1[1], 'bs')
plt.plot(testP2[0],testP2[1], 'rs')
plt.axis([0, 10, 0, 10])
plt.show()
```
